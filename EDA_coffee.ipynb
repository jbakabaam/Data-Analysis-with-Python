```python
import pandas as pd
```

```python
temp = pd.read_csv("data/coffee/소상공인시장진흥공단_상가(상권)정보_세종_202012.csv", sep='|',
                  encoding='utf-8')
temp
```

```python
# data 폴더에 있는 모든 csv 파일을 읽어오기 위해 glob을 이용하자
from glob import glob

# csv 목록 불러오기
file_names = glob("data/coffee/*.csv")
file_names
total = pd.DataFrame()

# 모든 csv 병합하기
for file_name in file_names:
    temp = pd.read_csv(file_name, sep='|', encoding='utf-8')
    total = pd.concat([total, temp])
```

```python
# reset index
total.reset_index(inplace=True)
total
```

```python
# 분석에 필요한 컬럼을 고른다
data = total[['상호명','지점명','상권업종대분류명','상권업종중분류명','시도명','시군구명',
             '행정동명']]
data
```

```python
### 전국 이디야 개수
df_ediya = df_coffee[df_coffee["상호명"].str.contains("이디야")]
df_ediya.index = range(len(df_ediya))
print("전국 이디야 개수 : ", len(df_ediya))
df_ediya
```

```python
### 시각화를 해보자
import seaborn as sns
import matplotlib.pyplot as plt
import platform

from matplotlib import font_manager, rc
%matplotlib inline
```

```python
### Mac 에서 사용가능한 한글 글꼴 확인코드
### Win10 에서는 아무 필요 없음
[f.name for f in font_manager.fontManager.ttflist if 'Neo' in f.name]
```

```python
### OS 폰트 (깨지지 않는 한글 글꼴)

path = "c:/Windows/Fonts/malgun.ttf"
if platform.system() == 'Darwin': # Mac
    font_name = 'Apple SD Gothic Neo'
    rc('font', family == 'Apple SaD Gothic Neo')
elif platform.system() == 'Windows': # Windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family = font_name)
else: # Linux
    font_name = font_manager.FontProperties(fname="/usr/share/fonts/nanumfont/NanumGothic.ttf")
    rc('font', family = "NanumGothic")
```

```python
# 주요 3대 브랜드 시각화

print("***** 전국 커피전문점 중 주요 3대 브랜드 입점 비율 *****")

starbucks_rate = (len(df_starbucks) / len(df_coffee) * 100)
twosome_rate = (len(df_2some) / len(df_coffee) * 100)
ediya_rate = (len(df_ediya) / len(df_coffee) * 100)

x = ["스타벅스", "투썸플레이스", "이디야"]
y = [starbucks_rate, twosome_rate, ediya_rate]

plt.figure(figsize=(6,6))
plt.title("주요 3대 커피브랜드 전국 입점 비율", fontdict={"fontsize" : 20})
sns.barplot(x=x, y=y)
plt.savefig("coffee_barplot.png")
plt.show()
```
