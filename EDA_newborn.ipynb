```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 깨짐 방지
```

```python
plt.plot(df.index, df['출생아 수'])
plt.plot(df.index, df['합계 출산율'])
```

```python
# Sub-Plot
fig, ax1 = plt.subplots(figsize=(10,7))
ax1.plot(df.index, df['출생아 수'], color='#ff812d')

ax2 = ax1.twinx() # x축 공유하는 axis
ax2.plot(df.index, df['합계 출산율'], color='#ffd100')
```

```python
# Sub-Plot (Bar)
fig, ax1 = plt.subplots(figsize=(10,7))
ax1.bar(df.index, df['출생아 수'], color='#ff812d')

ax2 = ax1.twinx() # x축 공유하는 axis
ax2.plot(df.index, df['합계 출산율'], color='#ffd100')
```

```python
# 추가
# Sub-Plot (Bar)
fig, ax1 = plt.subplots(figsize=(10,7))
ax1.set_ylabel('출생아 수(천명)')
ax1.bar(df.index, df['출생아 수'], color='#ff812d')

ax2 = ax1.twinx() # x축 공유하는 axis
ax2.set_ylabel('합계 출산율(가임여성 1명당)')
ax2.plot(df.index, df['합계 출산율'], color='#ffd100')
```

```python
# Graph More
fig.suptitle('출생아 수 및 합계 출산율') # 그래프 제목 추가
fig, ax1 = plt.subplots(figsize=(13,5)) # 그래프 크기 조정
ax1.set_ylabel('출생아 수(천명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index, df['출생아 수'], color='#ff812d')
for idx, val in enumerate(df['출생아 수']): # 데이터 표시
    ax1.text(idx, val+12, val, ha='center')

ax2 = ax1.twinx() # x축 공유하는 axis
ax2.set_ylabel('합계 출산율(가임여성 1명당)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index, df['합계 출산율'], color='#ffd100', marker='o', ms=10, lw=3, mec='w', mew=2) # 마커 추가 및 조정
for idx, val in enumerate(df['합계 출산율']):
    ax2.text(idx, val+ 0.09, val, ha='center')
```
