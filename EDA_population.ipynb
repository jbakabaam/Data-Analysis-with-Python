3. 필요한 데이터만 셀렉, 남자 기준

```python
df_m = pd.read_excel("/Users/jb/Desktop/python/population.xlsx", skiprows=3, index_col='행정기관', usecols='B,E:Y')
df_m.head()
```

4. 데이터 형변환(str->int)
---

```python
df_m.iloc[0].str.replace(',', '').astype(int)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int) # 실제 데이터로 치환
df_m
```

---

5. 필요한 데이터만 셀렉, 여자 기준
---

```python
df_w = pd.read_excel("/Users/jb/Desktop/python/population.xlsx", skiprows=3, index_col='행정기관', usecols='B,AB:AV')
df_w.head()
```

---

6. 컬럼명 동일하게 변경 (남자 컬럼명 = 여자 컬럼명) 및 형변환
---

```python
df_w.columns
```

```python
df_m.columns
```

```python
df_w.columns = df_m.columns
```

```python
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int)
df_w
```

---

7. Visualize
---

```python
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 깨짐 방지
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, df_m.iloc[0])
plt.title('인구 피라미드')
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, df_m.iloc[0] // 1000) # 단위: 1000명
plt.barh(df_w.columns, df_w.iloc[0]// 1000)
plt.title('인구 피라미드')
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000) # 단위: 1000명
plt.barh(df_w.columns, df_w.iloc[0]// 1000)
plt.title('인구 피라미드')
```
